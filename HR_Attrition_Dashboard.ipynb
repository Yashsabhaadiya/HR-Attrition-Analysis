```python
df = df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'])
print("Unnecessary columns dropped! ✅")
print("New Shape:", df.shape)
```

```python
print("Total Columns:", len(df.columns))
print("Total Rows:", df.shape[0])
print("Missing Values:", df.isnull().sum().sum())
print("Data is clean and ready! ✅")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title('Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df['Age_Group'] = pd.cut(df['Age'], 
                          bins=[18, 25, 35, 45, 55, 100], 
                          labels=['18-25', '26-35', '36-45', '46-55', '55+'])
plt.figure(figsize=(8,5))
sns.countplot(x='Age_Group', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by OverTime')
plt.xlabel('OverTime')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set2')
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Job Satisfaction')
plt.xlabel('Job Satisfaction (1=Low, 4=High)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
total = len(df)
left = df[df['Attrition'] == 'Yes'].shape[0]
rate = round(left / total * 100, 2)

print("=" * 40)
print("       HR ATTRITION - KEY INSIGHTS")
print("=" * 40)
print(f"Total Employees   : {total}")
print(f"Employees Left    : {left}")
print(f"Attrition Rate    : {rate}%")
print("=" * 40)
print("📌 Higher attrition in Sales department")
print("📌 Employees doing OverTime leave more")
print("📌 Low salary = higher attrition risk")
print("📌 younger employees leave more often")
print("📌 Low job satisfaction drives attrition")
print("=" * 40)
```
